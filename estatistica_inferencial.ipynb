
###  Introdução

Estatísticas inferenciais usam uma amostra aleatória dos dados coletados de uma população para descrever e fazer inferências sobre a
população. As estatísticas inferenciais são valiosas quando não é conveniente ou possível examinar cada membro de uma população inteira. 

Por exemplo, não seria prático medir o diâmetro de todos os pregos fabricados em uma fábrica, mas é possível medir o diâmetro de uma
amostra representativa de pregos e usar essas informações para fazer generalizações sobre os diâmetros dos pregos produzidos.

###  H0 e H1 (e seus erros)

De maneira geral são criadas e testadas as hipóteses H0 e H1 (Hipótese Nula e Hipótese Alternativa).

Se H0 for rejeitada, conclui-se que Beta1 é diferente de 0 e que existe uma relação estatisticamente significativa entre as duas variáveis.
Entretanto, se H0 não puder ser rejeitada, tem-se evidencias insuficientes para consluir que existe uma relação significativa.

Portanto, a intenção é rejeitar H0 para mostrar que há relação entre a variável Xi (variável independente) e y (variável dependente)

**Só lembrando**

**H0 = Hipótese Nula, aquela que queremos rejeitar.**

**H1 = Hipótese Alternativa, não queremos rejeitar.**


Uma vez que os testes de hipótese baseiam-se em informações de amostras, devemos admitir a possibilidade de **erros** na conclusão das hipóteses.

Conclusão Final do Teste de Hipótese: Aceitar ou Rejeitar H0

**Erro Tipo 1**: Rejeitar H0 quando H0 é verdadeira.

**Erro Tipo 2**: Aceitar H0 quando H0 for falsa.

###  Significância Estatística

Nível de Significância (α, alpha): é a probabilidade de comerter o erro do Tipo 1, portanto, ao selecionar alpha, controla-se a probabilidade de cometer o erro do Tipo 1. Se o custo de cometer um erro do Tipo 1 for alto, valores pequenos de alpha são preferíveis.

Mesmo podendo controlar a probabilidade de cometer o Erro do Tipo 1, nem sempre os testes de hipótese controlam a probabilidade de
cometer o Erro do Tipo 2. 

Portanto, se decidirmos **não rejeitar H0**, não podemos determinar quão confiantes podemos estar a respeito desta decisão.

Devido a incerteza associada a probabilidade de cometer um erro do Tipo 2 quando se realizam testes de significância, os estatísticos frequentemente recomendam que devemos usar a afirmação "**não rejeitar H0**" em vez de "**aceitar H0**".


###  Normalização

A normalização visa colocar os dados dentro de um padrão que facilite a sua análise. A normalização de dados Min-Max é um processo de 
transformar valores discrepantes usando uma escala que vai de 0,0 para o menor e 1,0 para o maior valor. Isso facilita a comparação de valores que 
foram medidos usando diferentes escalas, por exemplo, Celsius e Fahrenheit ou de diferentes unidades de medida como velocidade e distância.

###  P-Valor

P-Valor é a probabilidade de que a estatística de teste (como variável aleatória) tenha o valor extremo em relação ao valor observado quando H0 é verdadeira.

###  Teste de Hipótese Z

Teste Z é qualquer teste estatístico no qual a distribuição do teste sob a hipótese nula pode ser aproximada por uma distribuição normal.

É um teste estatístico usado para inferência (afirma a verdade de uma preposição em decorrência de sua ligação com outras já reconhecidas como
verdadeiras), capaz de determinar se a diferença entre a média da amostra e da população é grande o suficiente para ser significativa estatisticamente.

Esse teste é utilizado para amostras grande (> ou = 30)

Usa-se quando o desvio padrão é conhecido

Exemplo de como realizar:

1. Analise o problema
2. Determine a taxa de significância (α: 1%, 5% ou 10%)
3. Escreva as hipóteses (Meta: Rejeitar H0)
4. Calcule o Z-Score
5. Calcule o Z-Crítico (O limite de rejeição) e o P-Valor
6. Tome a decisão

###  Teste de Hipótese T

Adistribuição T possui características da distribuição normal:

Formato de sino

Simetria

Média igual a zero

Teste T pode ser conduzido para: 

Comparar uma amostra com uma população

Comparar duas amostras pareadas

Comparar duas amostras independentes

Aprincipal diferença é que a distribuição T acomoda “erros extras” para amostras pequenas (por isso seu formato mais arredondados nas extremidades).

O formato da distribuição T depende do tamanho da amostra, e consequentemente de graus de liberdade, quanto maior a amostra mais próxima
da Normal a distribuição T será. Com 30 ou mais observações a distribuição T é muito próxima da distribuição Normal.

Graus de liberdade = (n - 1) : número de informações independentes utilizadas para computar o desvio padrão amostral.
Quanto maior for o grau de liberdade, menor será a dispersão.

Conforme os graus de liberdade aumentam, a diferença entre a distribuição T e a distribuição Normal padrão vai ficando cada vez menor.
Para mais de 100 graus de liberdade, o valor normal padrão Z provê uma boa estimativa do valor T
